### Question 3

Create and analyze clusters of Toronto neighborhoods

```python
import folium
import numpy as np
import pandas as pd
import matplotlib.colors as colors
```

```python
toronto = folium.Map(location=[df["Latitude"][0], df["Longitude"][0]], zoom_start=10)
```

```python
def folium_graph(borough, neighbor, lat, long, color, map):
    for borough, neighbor, lat, long, color in zip(borough,neighbor,lat,long, color):
        
        colors_list = list(colors._colors_full_map.values())
        
        if color == 0:
            color = colors_list[0]
        else:
            color = colors_list[color]
        mark = folium.Popup('{}, {}'.format(neighbor, borough), parse_html=True)
        folium.CircleMarker(
            [lat, long],
            radius = 3,
            color=color,
            popup=mark,
            fill=True,
            fill_opacity=0.5,
            fill_color='#3366ff',
            parse_html=False
        ).add_to(map)
    
    return map
```

```python
folium_graph(df['Borough'],df['Neighbourhood'],df['Latitude'],df['Longitude'], np.zeros(len(df)), toronto)
```

### KMeans clustering

Use of Latitude and Longitude to cluster

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = 6
```

```python
clustering = pd.DataFrame({ 'Latitude' : df['Latitude'], 'Longitude' : df['Longitude'] })
```

```python
kmeans = KMeans(n_clusters=n_clusters,random_state=0).fit(clustering)
```

```python
# Add the clustering number to the dataframe
df.insert(0, 'Cluster', kmeans.labels_)
```

```python
toronto_clustered = folium.Map(location=[df["Latitude"][0], df["Longitude"][0]], zoom_start=10)
```

```python
folium_graph(df['Borough'],df['Neighbourhood'],df['Latitude'],df['Longitude'], df['Cluster'], toronto_clustered)
```
